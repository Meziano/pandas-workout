# Series
## Exercise 001
Create a series of 10 elements, random integers from 70 to 100, representing scores on a monthly exam. Set the index to be the month names, starting in September and ending in June. 
With this series, write code to answer the following questions:
* What is the student’s average test score for the entire year?
* What is the student’s average test score during the first half of the year (i.e., the first five months)?
* What is the student’s average test score during the second half of the year?
* Did the student improve their performance in the second half? If so, by how much?
## Workout
The exercise raises several questions:
* How do we define a series?  
  ```
  s = Series([10, 20, 30, 40, 50])
  ```
* How can we create 10 random integers from 70 to 100?
  The Python standard library’s random module has a randint method that returns a random integer:
  ```
  random.randint(0, 100) # this returns an integer between 0 and 100
  ```
  In NumPy, though:
  ```
   g = np.random.default_rng()  # create a new random-number generator object
   g.integers(70, 100, 10) # create an array of random integers
  ```
* How can we set the index of the series to month names?
  There are 2 possibilities to do that:
  - 1st possibility:
    ```
    g = np.random.default_rng(0)
    months = 'Sep Oct Nov Dec Jan Feb Mar Apr May Jun'.split()
    s = Series(g.integers(70, 101, 10),  index=months)
    ```
  - 2nd possibility:
    ```
    g = np.random.default_rng(0)    
    s = Series(g.integers(70, 101, 10))
    s.index = 'Sep Oct Nov Dec Jan Feb Mar Apr May Jun'.split()))
    ```


````python
import numpy as np
from pandas import Series
g = np.random.default_rng()
months = 'Sep Oct Nov Dec Jan Feb Mar Apr May Jun'.split()
s = Series(g.integers(70, 101, 10), index=months)
s
````

What is the student’s average test score for the entire year?

````python
s.mean()
print(f"The student's yearly average test score: {s.mean()}")
````

What is the student’s average test score during the first half of the year (i.e., the first five months)?

````python
# The student's average test score for the first half of the year (Sep to Feb) is:
first_half = s[:5]
print(f"The student's average test score for the first half of the year is: {first_half.mean()}")
````

````python
first_half_average = s.loc['Sep':'Jan'].mean()
print(f"The student's average test score for the first half of the year is: {first_half_average}")
````

What is the student’s average test score during the second half of the year?


````python
# The student's average test score during the second half of the year (Mar to Jun) is:
second_half = s[5:]
print(f"The student's average test score for the second half of the year is: {second_half.mean()}")
````

````python
second_half_average = s.loc['Feb':'Jun'].mean()
print(f"The student's average test score for the second half of the year is: {second_half_average}")
````

Did the student improve their performance in the second half? If so, by how much?

````python

print(f'Improvement: {second_half_average - first_half_average}')
````

````python
import math
var = math.sqrt(((s - s.mean()) ** 2).sum() / s.count())
print(f"1. Variance: {var}, s.std(): {s.std()}")
print(f"2. Variance: {var}, s.std(): {s.std(ddof=0)}")
````
