# EXERCISE: 3 Counting tens digits
In this exercise, I want you to generate 10 random integers in the range 0 to 100.  
(Remember that the np.random.randint function returns numbers that include the lower bound but exclude the upper bound.)  
Create a series containing those numbers’ tens digits. Thus, if our series contains 10, 25, 32, we want the series 1, 2, 3.

# Working it out
Series generated with np.random.randint(0, 100, 10), will all range from 0 (at the low end) to 99 (at the high end).   
To get the `tens digit`:
* we divide the series by 10. That turns the `dtype` into `floating type`and moves the decimal point one position to the left 
* we turn the series back to `np.int` that remove the fractional part of the number.  
```
print(f"s:\n{s}, \n(s/10).astype(int) :\n{(s/10).astype(int)}")
```
A better way is to use Python’s `//` **floordiv** operator which performs an integer division.   
If we divide the series by 10 using `//`, we’ll still get our `dtype` of `int`. 
```
print(f"s:\n{s}, \ns //10 :\n{s//10}")
```

Another way is to convert the series values to strings with `s.astype(str)` and then use `get()` to extract a specific digit.   
`get(i)` returns the character at position `i`, counting from the left.   
For example, if  testStr = '5891420' :
* testStr.get(0)  returns  5
* testStr.get(1)  returns  8
* ..
* testStr.get(6)  returns  0

You can also use negative indexes to count from the end (from the right):
* testStr.get(-1)  returns  0
* testStr.get(-2)  returns  2
* ..
* testStr.get(-7)  returns  NaN 

To get the tens digit we use s.get(-2) and if the reulst is `NaN` we use the `fillna` method to replace it with '0':
```
s.astype(str).get(-2).fillna(`0`)
```

# Solution
import numpy as np
from pandas import Series
g = np.random.default_rng()

s = Series(g.integers(0, 101, 10))
print(f"s:\n{s}, \ns //10 :\n{s//10}")
print(f"\n\ns.astype(str).str.get(-2).fillna('0'): \n{s.astype(str).str.get(-2).fillna('0').astype}")
